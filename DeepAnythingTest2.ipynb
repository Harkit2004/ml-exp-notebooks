```python
from transformers import pipeline
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

pipe = pipeline(task="depth-estimation", model="depth-anything/Depth-Anything-V2-Small-hf")
```

```python
image = Image.open('demo.jpg')
depth = pipe(image)["depth"]
```

```python
depth_gray = depth / np.max(depth) * 255
depth_gray = depth_gray.astype(np.uint8)

plt.imshow(depth_gray, cmap='gray')
plt.colorbar()
plt.show()
```

```python
plt.imsave("depth_heightmap.png", depth_gray, cmap='gray')
```

```python
scad_content = """filename = "/content/depth_heightmap.png";
center_model = true;
invert_heightmap = false;
height_scale = 1.5;
convexity_hint = 10;

scale([1, 1, height_scale])
surface(file = filename, center = center_model, invert = invert_heightmap, convexity = convexity_hint);
"""

r.render(scad_content, outfile='output.stl')
```
